```python
url_1 = '/content/SaoPaulo_OnlyAppartments.csv'
url_2 = '/content/SaoPaulo_OnlyAppartments_2024-11-25.csv'

df1 = pd.read_csv(url_1)
df2 = pd.read_csv(url_2)
```

```python
df2.drop(columns=['Longitude','Latitude'], inplace=True)
df2.columns
```

```python
df = pd.concat([df1, df2], ignore_index=True)
```

```python
df.info()
```

```python
df['Endereco_limpo'] = df['Adress'].str.split('-').str[0].str.strip()
df.head()
```

```python
df[['rua_numero', 'bairro']] = df['Endereco_limpo'].str.rsplit(',', n=1,expand=True)
df.head()
```

```python
df['rua'] = df['rua_numero'].str.extract(r'([A-Za-z\s]+)')
df.head()
```

```python
df['cidade'] = df['Adress'].str.split('-').str[1].str.strip()
df.head()
```

```python
df['cidade'] = df['cidade'].str.split('/').str[0].str.strip()
df.head()
```

```python
df.drop(columns=['Adress','Endereco_limpo','rua_numero'], inplace=True)
df.head()
```

```python
df['bairro'] = df['bairro'].str.strip()
```

```python
df['data_criacao'] = df['created_date'].str.split('T').str[0].str.strip()
df.head()
```

```python
df['data_criacao'] = pd.to_datetime(df['data_criacao'])
df.head()
```

```python
df['ano_criacao'] = df['data_criacao'].dt.year
df.head()
```

```python
df.drop(columns=['extract_date','created_date', 'ID'], inplace=True)
df.head()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
df.rename(columns={ 'Price': 'preco', 'below_price':'abaixo_media_preco', 'Area':'area' ,'Bedrooms': 'qtde_quartos','Bathrooms':'qtde_banheiros', 'Parking_Spaces':'qtde_vagas_estac', 'ID':'id'}, inplace=True)
df.head()
```

```python
df.drop(df[df['preco'] == 0].index, inplace=True)
df.head()
```

```python
(df['preco'] == 0).value_counts()
```

deletar cidades diferentes


```python
cidades_sp = [
    'São Paulo', 'São Bernardo do Campo', 'Santo André', 'Osasco', 'Piracicaba',
    'Barueri', 'Itapecerica da Serra', 'Cosmópolis', 'Itaquaquecetuba', 'Cotia',
    'Poá', 'Arujá', 'Campinas', 'Guarulhos', 'Salto', 'Sertãozinho', 'Guarujá',
    'Taboão da Serra', 'Diadema', 'Jacareí', 'Taubaté'
]

df = df[df['cidade'].isin(cidades_sp)].reset_index(drop=True)
```

```python
df['cidade'].unique()
```

deletar bairros e cidades vazias


```python
df.drop(df[df['bairro'].isnull()].index, inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
df['valor_m2'] = (df['preco'] / df['area']).round(2)
df.head()
```

```python
df_graficos = df.groupby('bairro')['valor_m2'].agg(['mean', 'count'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df_graficos['mean'], bins=20, kde=True)
plt.title('Histograma da Média do Valor/m² por Bairro')
plt.show()
```

```python
# Boxplot (distribuição da média do valor/m²)
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_graficos['mean'])
plt.title('Boxplot da Média do Valor/m² por Bairro')
plt.show()
```

```python
bairros_caros = df_graficos.nlargest(10, 'mean')
plt.figure(figsize=(26, 8))
sns.barplot(x=bairros_caros.index, y=bairros_caros['mean'])
plt.title('Top 10 Bairros Mais Caros')

```

```python
bairros_baratos = df_graficos.nsmallest(10, 'mean')
plt.figure(figsize=(30, 8))
sns.barplot(x=bairros_baratos.index, y=bairros_baratos['mean'])
plt.title('Top 10 Bairros Mais Baratos')

```
